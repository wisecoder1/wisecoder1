```python
vegetables = ['potato','carrot','ginger','red chilly','lady finger']
foods = ['bread','snack','burger','cake']
farming = ['rise','gram','corn','millet']

#Vegetable items
veg1 =vegetables [0]
veg2 =vegetables [1] 
veg3 =vegetables [2]
veg4 =vegetables [3]
#food items
food1 =foods [0]
food2 =foods [1]
food3 =foods [2]
food4 =foods [3]
#farming items
f1 =farming [0]
f2 =farming [1]
f3 =farming [2]
f4 =farming [3]

#vegitable items price
veg_item1=33
veg_item2=22
veg_item3=40
veg_item4=11    

#food items price
f_item1 = 15
f_item2 = 30   
f_item3 = 50    
f_item4 = 20 

#farming items price
far_item1 = 50  
far_item2 = 20  
far_item3 = 23  
far_item4 = 40  

header = '\tWelcome to online shopping '
user_name = str(input("Enter Your Name. "))
print(header+user_name)

while True:
    user_address = str(input("Enter Your Address. "))
    if len(user_address)<=30:
        print('Your Address must be greater than 40 Charactor ')
        continue
    else:
        print('Your Address has successfully added! ')
        break 

while True:
    print(" Get Started s? ")
    user_input1 = str(input('yes / no. '))
    if user_input1=='yes':
        print('\tPRODUCTS.')
        print('\nVegetable category.')
        print('')
        print("Price of {} is: ${:.2f}".format(veg1,veg_item1))
        print("Price of {} is: ${:.2f}".format(veg2,veg_item2))
        print("Price of {} is: ${:.2f}".format(veg3,veg_item3))
        print("Price of {} is: ${:.2f}".format(veg4,veg_item4))
        while True:
            select_product1 = str(input('Select Vegetable: '))
            if select_product1 not in vegetables:
                print(">>>Please fill valid product ")
                continue
            else:
                print('Your product has successfully added.',select_product1)
                if select_product1=='potato':
                    pay = veg_item1
                elif select_product1=='carrot':
                    pay = veg_item2
                elif select_product1=='ginger':
                    pay = veg_item3
                else:
                    pay = veg_item4
                break
        print('')
        print('\tSEE MORE>>>')
        print('\nFood category.')
        print('')
        print("Price of {} is: ${:.2f}".format(food1,f_item1))
        print("Price of {} is: ${:.2f}".format(food2,f_item2))
        print("Price of {} is: ${:.2f}".format(food3,f_item3))
        print("Price of {} is: ${:.2f}".format(food4,f_item4))
        while True:
            select_product2 = str(input('Select food'))
            if select_product2 not in foods:
                print(">>>Please fill valid product ")
                continue
            else:
                print('Your product has successfully added.', select_product2,)
                if select_product2==food1:
                    p2_pay =  f_item1
                elif select_product2==food2:
                    p2_pay = f_item2
                elif select_product2==food3:
                    p2_pay = f_item3
                else:
                    p2_pay = f_item4
                break
                
        print('\tSEE MORE>>> ')
        print('\nFood farming.')
        print('')
        print("Price of {} is: ${:.2f}".format(f1,far_item1))
        print("Price of {} is: ${:.2f}".format(f2,far_item2))
        print("Price of {} is: ${:.2f}".format(f3,far_item3))
        print("Price of {} is: ${:.2f}".format(f4,far_item4))
        while True:
            select_product3 = str(input('Select food'))
            if select_product3 not in farming:
                print(">>>Please fill valid product ")
                continue
            else:
                print('Your product has successfully added.', select_product3)
                if select_product3==f1:
                    p3_pay = far_item1
                elif select_product3==f2:
                    p3_pay = far_item2
                elif select_product3==f3:
                    p3_pay = far_item3
                else:
                    p3_pay = far_item4
                break
        price = pay+p2_pay+p3_pay
        print("Your Added Items press Enter to  Order! ")
        print(select_product1, '' ,select_product2, '' ,select_product3,'','Total Price ${:.2f}'.format(price))
        str(input('Press Enter '))
        print('Your Order Successfully Done on your Address ')
        break
```
